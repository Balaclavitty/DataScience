```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
cars = sns.load_dataset('mpg').dropna()

cars.shape
```

```python
cars.head(4)
```

```python
sns.set_style('whitegrid')
```

```python
cars.cylinders.value_counts()
```

```python
cars = cars[cars.cylinders.isin([4, 6, 8])]
```

```python
sns.swarmplot(cars.horsepower)
```

```python
sns.swarmplot(x=cars.origin, y=cars.horsepower)
```

```python
sns.swarmplot(x='origin', y='horsepower',
             data=cars
             )
```

```python
sns.swarmplot(x='origin', y='horsepower',
              hue='cylinders',
             data=cars
             ) #adding categories for better understanding
```

```python
sns.swarmplot(x='origin', y='horsepower',
              hue='cylinders',
             data=cars, dodge=True
             ) #dodge allows to split the data to be shown side by side instead all in one place 
#dodge not really recommended for large datasets, rather their samples
```

```python
usa = cars[cars.origin=='usa']
```

```python
sns.boxplot(x=usa.cylinders, y=usa.horsepower)
sns.swarmplot(x=usa.cylinders, y=usa.horsepower) #boxplot adn swarmplot overlap
```

```python
sns.boxplot(x=usa.cylinders, y=usa.horsepower)
sns.swarmplot(x=usa.cylinders, y=usa.horsepower,
             color='black') #different color allows for better distinction
```

```python
sns.boxplot(x=usa.cylinders, y=usa.horsepower)
sns.swarmplot(x=usa.cylinders, y=usa.horsepower,
             color='black', alpha=0.5) #alpha paramter enables transparency for the dots in swarmplot 
```

```python
sns.boxplot(x=usa.cylinders, y=usa.horsepower,
           whis=np.inf) #whiskers elongated to the outliers
sns.swarmplot(x=usa.cylinders, y=usa.horsepower,
             color='black', alpha=0.5)
```

```python
sns.violinplot(x=usa.cylinders, y=usa.horsepower)
sns.swarmplot(x=usa.cylinders, y=usa.horsepower)
```

```python
sns.violinplot(x=usa.cylinders, y=usa.horsepower,
              scale='width') #adjusting to the same width, so all the white plots in swarmplot are being shown
sns.swarmplot(x=usa.cylinders, y=usa.horsepower,
             color='white')
```

```python
sns.violinplot(x=usa.cylinders, y=usa.horsepower,
              scale='width') 
sns.swarmplot(x=usa.cylinders, y=usa.horsepower,
             color='white', size=4) #size allows for dot size manipulation
```

```python
sns.violinplot(x=usa.cylinders, y=usa.horsepower,
              scale='width', inner=None) #inner=None allows for switching of the box in violinplots
sns.swarmplot(x=usa.cylinders, y=usa.horsepower,
             color='white', size=3)
```

```python
sns.violinplot(x=usa.cylinders, y=usa.horsepower,
              scale='width', inner=None, color='xkcd:forest green') #changing violinplot colors  
sns.swarmplot(x=usa.cylinders, y=usa.horsepower,
             color='white', size=3)
```

```python
sns.swarmplot(cars.horsepower, linewidth=1, edgecolor='black') #changed linewidth for the outline together with outline color
```

```python
plt.figure(figsize=(8, 4))

sns.swarmplot(cars.horsepower, marker='*') #marker changes the shape of the marker
```
