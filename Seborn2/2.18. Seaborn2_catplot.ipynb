```python
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg').dropna()
cars = cars[cars.cylinders.isin([4, 6, 8])]
cars['type'] = ['old' if x <=76 else 'new' for x in cars.model_year] #adds old/new column
```

```python
cars.head()
```

```python
cars.shape
```

```python
sns.set_style('white')
```

```python
sns.catplot(x='horsepower', data=cars)
```

```python
sns.catplot(y='horsepower', data=cars) #vertical catplot
```

```python
sns.catplot(x='horsepower', data=cars, kind='box')
```

```python
sns.catplot(x='horsepower', data=cars, kind='violin')
```

```python
sns.catplot(x='origin', y='horsepower', 
            data=cars, 
            kind='box')
```

```python
sns.catplot(x='origin', y='horsepower', 
            data=cars,
            hue='cylinders',
            kind='box') #grouped by cylinders and origin
```

```python
sns.catplot(x='origin', y='horsepower', 
            data=cars,
            hue='cylinders',
            kind='violin')
```

```python
sns.set_style('darkgrid')
```

```python
sns.catplot(x='cylinders', y='horsepower', 
            data=cars,
            kind='strip',
           col='origin') #assigning origin to column argument allows to present data in 3 different graphs (-> 3 different countries)
```

```python
sns.catplot(x='cylinders', y='horsepower', 
            data=cars,
            kind='strip',
           col='origin',
           row='type') #columns split in countries, rows in type
```

```python
sns.catplot(x='cylinders', y='horsepower', 
            data=cars,
            kind='violin',
           col='origin',
           row='type', 
           row_order=['new', 'old']) #row_order allows for re-arranging the order the graphs are shown
```

```python
sns.catplot(x='cylinders', y='horsepower', 
            data=cars,
            kind='box',
           col='origin',
           palette='winter',
           whis=1) #smaller whiskers allows to show more outliers
```

```python
sns.catplot(x='cylinders', y='horsepower', 
            data=cars,
            hue='origin',
            kind='strip',
           col='origin',
           palette='winter',
           s=24, 
           alpha=0.3, #transparency
           linewidth=2) 
```

```python
g = sns.catplot(x='cylinders', y='horsepower',
               data=cars,
               col='origin',
               kind='box')
```
