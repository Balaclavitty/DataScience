```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
df['decade'] = [age//10*10 for age in df.age] #creates new column decade, which takes age and rounds it down to provide decade
df.head()
```

```python
sns.set_style('white')
```

```python
sns.relplot(x='age', y='fare', data=df)
```

```python
sns.relplot(x='age', y='fare',
           data=df,
           hue='pclass',
           style='pclass',
           size='pclass') 
```

```python
sns.relplot(x='age', y='fare',
           data=df,
            kind='line') #lineplot
```

```python
sns.relplot(x='decade', y='fare',
           data=df,
            kind='line',
           hue='pclass',
           style='pclass')
```

```python
sns.relplot(x='age', y='fare',
          data=df,
          col='pclass') #1st class shows a lot of variety compared to 2nd and 3rd
```

```python
sns.relplot(x='decade', y='fare',
            data=df,
            kind='line',
            col='pclass')
```

```python
sns.relplot(x='age', y='fare',
            data=df,
            col='deck',
            col_wrap=3) #wraping plots every 3rd
```

```python
sns.relplot(x='age', y='fare',
            data=df,
            col='pclass',
            hue='survived',
            s=150,
            palette=['black', 'red'],
            markers=['X', 'o'], #markers can be manually assigned by using symbols in list format
            style='survived')
```

```python
g = sns.relplot(x='decade', y='survived',
           data=df[df.age < 70], #excluding over 70 years old 
           kind='line',
           hue='sex',
           col='pclass',
           errorbar=None) #switching off confidence intervals

g.set_xlabels('Passenger Age Group') #setting labels on g.FacetGrid
g.set_ylabels('Percent Survived')

#adding to FacetGrid allows to update it accordingly
```
