```python
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg').dropna()
cars.head()
```

```python
sns.set_style('white')
```

```python
sns.displot(x='weight', data=cars, kde=True) #adding kde line
```

```python
sns.displot(x='weight', data=cars, kind='kde')
```

```python
sns.displot(x='weight', data=cars, kind='ecdf', rug=True) #rugplots allow to visualize density and distribution when dealing with small sets
```

```python
sns.displot(x='weight', data=cars, hue='origin')
```

```python
sns.displot(x='weight', data=cars, kind='kde', hue='origin')
```

```python
sns.displot(x='weight', y='horsepower', data=cars)
```

```python
sns.displot(x='weight', y='horsepower', data=cars, kind='kde', rug=True)
```

```python
sns.displot(x='weight', data=cars, hue='origin', col='origin') #FacetGrid properties
```

```python
sns.displot(x='weight', data=cars, col='origin', kind='kde', rug=True)
```

```python
sns.displot(x='weight', data=cars[cars.cylinders.isin([4, 6, 8])],
           col='origin',
           row='cylinders')
```

```python
sns.displot(x='weight', data=cars, hue='origin', col='origin', palette='crest')
```

```python
sns.displot(x='weight', data=cars, hue='origin', col='origin', height=2) #updating size
```

```python
sns.displot(x='weight', data=cars, hue='origin', col='origin', height=2, aspect=1.5) #updating the width of teh charts
```

```python
sns.displot(x='weight', data=cars, bins=20)
```

```python
sns.displot(x='weight', data=cars, kind='kde', fill=True, lw=3)
```

```python
sns.displot(x='weight', data=cars, kind='kde', rug=True, rug_kws={'height': 0.05})
#rug_kws (rug keywords allow to adjust rugplot)
```

```python
sns.displot(x='weight', data=cars, col='origin', facet_kws={'despine': False}) #compare the frames of the charts with above
```

```python
g = sns.displot(x='weight', data=cars, col='origin')
```

```python
g.axes_dict.items()
```

```python
g = sns.displot(x='weight', data=cars, col='origin')

for col_val, ax in g.axes_dict.items():
    if col_val == 'europe':
        ax.set_facecolor("#E2F2FF")
#when orgin is equal to europe the facecolor is changed to "#E2F2FF"
```
