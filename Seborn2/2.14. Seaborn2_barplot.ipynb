```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins.head()
```

```python
penguins.shape
```

```python
sns.set_style('darkgrid')
```

```python
sns.barplot(x='species', y='body_mass_g', data=penguins)
```

```python
penguins.groupby('species').body_mass_g.mean() #barplot takes a mean for species
```

```python
sns.barplot(x='body_mass_g', y='species', data=penguins) #horizontal barplot
```

```python
sns.barplot(x='species', y='body_mass_g', 
            data=penguins,
            errorbar=('ci', 95)) #confidence interval by default is 95 -> przedział ufności
```

```python
sns.barplot(x='species', y='body_mass_g', 
            data=penguins,
            errorbar=('ci', 68)) #reducing manually confidence interval
```

```python
sns.barplot(x='species', y='body_mass_g', 
            data=penguins,
            errorbar=None) #switching off confidence interval
```

```python
sns.barplot(x='species', y='body_mass_g', 
            data=penguins,
            estimator=np.std) #switching the estimator to show standard deviation istead of confidence interval
```

```python
penguins.groupby('species').body_mass_g.describe()
```

```python
sns.barplot(x='species', y='body_mass_g', 
            data=penguins,
            estimator=np.max)
```

```python
sns.barplot(x='species', y='body_mass_g', 
            data=penguins,
            hue='sex') #grouping also by sex of the species
```

```python
sns.barplot(x='species', y='body_mass_g', 
            data=penguins,
            order=['Gentoo', 'Chinstrap', 'Adelie']) #switching the order manually
```

```python
sns.barplot(x='species', y='body_mass_g',
            hue='sex',
            order=['Gentoo', 'Chinstrap', 'Adelie'],
            hue_order=['Female', 'Male'],
            data=penguins) #hue_order also can be manually adjusted 
```

```python
sns.barplot(x='species', y='body_mass_g', 
            data=penguins,
            color='pink')
```

```python
sns.barplot(x='species', y='body_mass_g', 
            data=penguins,
            color='pink',
            errcolor='red',
            errwidth=8) #things adjusted: error width, error color, color of the bars
```

```python
sns.barplot(x='species', y='body_mass_g', 
            data=penguins,
            palette='twilight',
            edgecolor='black',
            lw=4) #things adjusted palette, linewidth, edgecolor
```
