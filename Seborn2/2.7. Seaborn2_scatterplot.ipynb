```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
diamonds = sns.load_dataset('diamonds')

diamonds.shape
```

```python
diamonds.head()
```

```python
diamonds_clean = diamonds[
            diamonds.cut.isin(['Premium', 'Good']) &
            diamonds.color.isin(['D', 'F', 'J'])].sample(n=100, random_state=22)

diamonds_clean.shape
```

```python
sns.set_style('dark')
```

```python
sns.scatterplot(x=diamonds_clean.carat, y=diamonds_clean.price)
```

```python
sns.scatterplot(x='carat', y='price', data=diamonds_clean) #different syntax, same result
```

```python
#sns.scatterplot(x='x', y='y', data=diamonds)
```

```python
sns.scatterplot(x='carat', y='price', hue='cut',
                data=diamonds_clean
               ) #allows to show scatterplot in relation to the cut variable 
```

```python
sns.scatterplot(x='carat', y='price', hue='cut',
                data=diamonds_clean,
                palette=['purple', '#55CCCC']
               ) #reuqires more colors since more categories
```

```python
sns.scatterplot(x='carat', y='price', hue='depth',
               data=diamonds_clean
               ) #darker color represents higher depth
```

```python
sns.scatterplot(x='carat', y='price', hue='depth',
               data=diamonds_clean,
                s=100
               ) #s controls size of the dots
```

```python
sns.scatterplot(x='carat', y='price', style='cut',
               data=diamonds_clean,
                s=100
               ) #allows the change the style of plotting 
```

```python
sns.scatterplot(x='carat', y='price', size='cut',
               data=diamonds_clean,
               ) #adjusts the size of the dots according to the legend 
```

```python
sns.scatterplot(x='carat', y='price', size='cut',
               data=diamonds_clean,
                sizes=[150, 50]
               ) #manually adjusting sizes of the dots
```

```python
sns.scatterplot(x='carat', y='price', size='depth',
               data=diamonds_clean,
               )
```

```python
sns.scatterplot(x='carat', y='price', 
                hue='cut',
                style='cut',
               data=diamonds_clean,
                s=100
               )
```

```python
sns.scatterplot(x='carat', y='price', 
                hue='cut',
                style='color',
               data=diamonds_clean,
                s=100
               )
```

```python
sns.scatterplot(x='carat', y='price',
               data=diamonds_clean,
                s=100,
                alpha=0.6
               ) #alpha adds transparency
```

```python
sns.scatterplot(x='carat', y='price',
               data=diamonds_clean,
                s=200,
                marker='*'
               ) #switches to using stars
```

```python
sns.scatterplot(x='carat', y='price',
               data=diamonds_clean,
                s=200,
                marker='*',
                edgecolor='black'
               ) #edge color allows changing outline of the symbol
```
