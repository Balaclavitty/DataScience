```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
cars = sns.load_dataset('mpg')
```

```python
cars.head()
```

```python
cars.groupby('origin').cylinders.value_counts()
```

```python
origin_cyl =(cars
    .groupby('origin')
    .cylinders
    .value_counts()
    .unstack()
    .fillna(0)
) #unstacking and filling out the NaN with 0
```

```python
sns.heatmap(origin_cyl) #heatmap created on above values: 
```

```python
test_array = np.array([[1, 2], [3, 4]])

test_array
```

```python
sns.heatmap(test_array)
```

```python
origin_cyl.T #inverts the matrix
```

```python
sns.heatmap(origin_cyl.T)
```

```python
column_to_delete = ['origin', 'name'] #dropping non-dummified columns
cars2 = cars.drop(column_to_delete, axis=1)

sns.heatmap(cars2.corr())
```

```python
sns.heatmap(origin_cyl, cmap='Blues')
```

```python
sns.heatmap(cars2.corr(), cmap='RdBu')
```

```python
sns.heatmap(cars2.corr(), cmap='RdBu', center=0) #changing the center to exact 0 
```

```python
sns.heatmap(cars2.corr(), cmap='RdBu', vmin=-1, vmax=1) #changing the scale to exact -1 and 1 
```

```python
sns.heatmap(origin_cyl, cmap='Blues', annot=True) #adding annotations (numbers in the squares)
```

```python
sns.heatmap(origin_cyl, cmap='Blues', annot=True, fmt=".0f") #fmt set to .0f allows to set annotions to integer, .2f -. 2-decimal points
```

```python
sns.heatmap(origin_cyl, 
            cmap='Blues', 
            annot=True, 
            fmt=".0f",
           annot_kws={
               'fontsize': 16,
               'fontweight': 'bold',
               'fontfamily': 'serif'
           } #annot_kws allows for manual adjustments for the text inside squares
           )
```

```python
sns.heatmap(cars2.corr(), cmap='RdBu', linewidth=1, linecolor='black') #lines added to very square, linecolor can also be adjusted
```

```python
labels = ['MPG', 'Cylinders', 'Displacement', 'Horsepower', 'Weight', 'Acceleration', 'Model Year']
```

```python
sns.heatmap(cars2.corr(), 
            cmap='RdBu',
            xticklabels=labels,
            yticklabels=labels
           ) #adjusting manuallu labels
```

```python
sns.heatmap(cars2.corr(), 
            cmap='RdBu',
            xticklabels=False, #switching off labels for x-values
            yticklabels=labels
           ) 
```

```python
sns.heatmap(cars2.corr(), 
            cmap='RdBu',
            square=True
           ) #changing to shape to perfect squares
```
