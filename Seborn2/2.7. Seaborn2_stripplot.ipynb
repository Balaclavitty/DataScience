```python
cars = cars[cars.cylinders.isin([4,6,8])]
```

```python
sns.stripplot(cars.weight)
```

```python
sns.stripplot(x=cars.weight, y=cars.origin)
```

```python
sns.stripplot(x='weight', y='origin',
             data=cars
             )
```

```python
sns.stripplot(x='weight', y='origin',
              hue='cylinders',
             data=cars
             )
```

```python
sns.stripplot(x='weight', y='origin',
              hue='cylinders',
             data=cars, dodge=True #dogde=True allows for separete stripplots 
             )
```

```python
sns.stripplot(y='weight', x='origin',
             data=cars
             ) #changed directions
```

```python
sns.stripplot(x=cars.weight, y=cars.origin, jitter=False) #removes granularity 
```

```python
sns.stripplot(x=cars.weight, y=cars.origin, jitter=0.05) #jitter can be also used to adjust granularity 
```

```python
import ipywidgets
from IPython.display import display, clear_output

def jitter_widget(jitter=0.1):
    clear_output(wait=True)
    sns.stripplot(x=cars.weight, y=cars.origin, jitter=jitter)
    plt.show()
```

```python
ipywidgets.interact(jitter_widget, jitter=(0, 0.65, 0.05))
```

```python
sns.stripplot(x=cars.weight, y=cars.origin, alpha=0.5) #alpha regulates transparency
```

```python
sns.stripplot(x=cars.weight, y=cars.origin, alpha=0.5,
             size=3) #size allows for dot size manipulation
```

```python
sns.stripplot(x=cars.weight, y=cars.origin, linewidth=1, edgecolor='black')
```

```python
sns.stripplot(x=cars.weight, y=cars.origin,
             alpha=0.7, size=10, jitter=0.25, marker='*')
```
