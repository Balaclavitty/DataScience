```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.set_style('darkgrid')
```

```python
sns.pairplot(tips) #labels correspond to scatter not histograms
#default setup has histograms on diagonal and scatter off diagonals
```

```python
tips.dtypes
```

```python
sns.pairplot(tips, diag_kind='kde') #changing diagonal to kde plots
```

```python
sns.pairplot(tips, kind='reg') #changing of diagonal
```

```python
g = sns.pairplot(tips)
```

```python
type(g)
```

```python
g = sns.pairplot(tips) #grid can be saved as an object and added on the plots, refer to kde 
g.map_upper(sns.kdeplot, n_levels=6, color='xkcd:wine red')
```

```python
tips['weekend'] = tips.day.isin(['Sat', 'Sun'])
```

```python
tips.head()
```

```python
sns.pairplot(tips, hue='weekend') #comparing weekday and weekend data
```

```python
sns.pairplot(tips, hue='weekend', vars=['total_bill', 'tip']) #now pairplot takes only total_bill and tip 
```

```python
sns.pairplot(tips,
            hue='weekend',
            y_vars=['tip'],
            x_vars=['total_bill', 'size'],
            kind='reg',
            height=4) #specifing display of tip of only y_axis, x_axis total_bill and size in reference to tip
```

```python
from ipywidgets import interactive 

def size_widget(height=2.5, aspect=1):
    sns.pairplot(tips, height=height, aspect=aspect) #moving height changes size of the pairplot, aspect impacts width
```

```python
interactive(size_widget, height=(1, 3.5, 0.5), aspect=(0.5, 2, 0.25))
```

```python
sns.pairplot(tips, hue='day', palette='plasma')
```

```python
tips2 = tips.drop('weekend', axis=1)
tips2.head()
```

```python
sns.pairplot(tips2, diag_kws={'color': 'gray', 'alpha': 0.6}) #diagonal keywords allow to change parameters for diagonal plots
```

```python
sns.pairplot(tips2,
            kind='reg',
            plot_kws={'ci': None, 'color': 'xkcd:salmon',
                     'scatter_kws': {'color': 'grey'}})
```
