```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
diamonds = sns.load_dataset('diamonds')
diamonds.shape
```

```python
diamonds.head()
```

```python
diamonds = diamonds.sample(n=200, random_state=44)

diamonds.shape
```

```python
sns.set_style('dark')
```

```python
sns.regplot(x=diamonds.carat, y=diamonds.price)
```

```python
sns.regplot(x='carat', y='price', data=diamonds)
```

```python
sns.regplot(x='carat', y='price', data=diamonds,
           fit_reg=False) #without the line inbetween 
```

```python
sns.regplot(x='carat', y='price', data=diamonds,
           scatter=False) #turining off the scatter
```

```python
sns.regplot(x='carat', y='price', data=diamonds,
           ci=None) #ci set to None switches of the confidence intervals
```

```python
cut_map = {
    'Fair': 1,
    'Good': 2,
    'Very Good': 3,
    'Premium': 4,
    'Ideal': 5
}

diamonds['cut_value'] = diamonds.cut.map(cut_map)
```

```python
diamonds.cut_value.value_counts()
```

```python
is_ordered = diamonds['cut_value'].cat.ordered

if not is_ordered:
    diamonds['cut_value'] = diamonds['cut_value'].cat.as_ordered()
```

```python
import numpy as np
```

```python
sns.regplot(x='cut_value', y='price', data=diamonds,
           x_estimator=np.mean) #estimator changed to mean
```

```python
sns.regplot(x='carat', y='price', data=diamonds,
           order=2) #order allows to print polynomial regression 

#Aprokysmacja wielomianowa(polynomial regression): to model the relationship between a dependent variable (target) and one or more independent variables (predictors) by fitting a polynomial equation to the data. In polynomial regression, the relationship between the variables is represented by a polynomial equation of a specified degree.
```

```python
x_example=[0, 0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5]
y_example=[0.1, 0.8, 2.2, 2.7, 3.8, 4.5, 6.2, 6.8, 7.9, 9.4, 30.4]
```

```python
sns.regplot(x=x_example,
           y=y_example,
           ci=None)
```

```python
sns.regplot(x=x_example,
           y=y_example,
           ci=None,
           robust=True)
#robust regression is designed to be less sensitive to outliers
```

```python
sns.set_style('white')
```

```python
sns.regplot(x='carat', y='price', data=diamonds,
           marker='d') #diamond shaped markers
```

```python
sns.regplot(x='carat', y='price', data=diamonds,
           scatter_kws={'s':100, 'alpha':0.5, 'color':'lightgray'}) #scatter_kws allows to pass all properties that we want to change
```

```python
sns.regplot(x='carat', y='price', data=diamonds,
           ci=None,
           line_kws={'lw': 4, 'color': 'black', 'linestyle':'-.'}) #line_ws allows to change properties of the line
```
