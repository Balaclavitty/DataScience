# Empirical cumulative distribution function (dystrbuanta empiryczna)

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.shape
```

```python
sns.set_style('darkgrid')
```

```python
sns.ecdfplot(x='tip', data=tips)
#ECDF shows the value of the data and its proportion to the overall set (exmaple: 1 number in 10 dataset represents 0.1 ) 
```

```python
tips.tip.min()
```

```python
tips.tip.max()
```

```python
tips.tip.value_counts().head() # tip 2$ was most common (33 times)
```

```python
sns.ecdfplot(x='tip', data=tips)
plt.axvline(4, c='red') #by the time it reaches 4$, 80% of the data was already seen
```

```python
sns.ecdfplot(y='tip', data=tips) #proportion on y-axis
```

```python
sns.ecdfplot(y='tip', data=tips)
plt.axvline(0.5, c='red') #50% of data 
```

```python
tips.tip.median()
```

# Additional options

```python
sns.ecdfplot(x='tip', data=tips, hue='time') #dinner tips are larger
```

```python
sns.ecdfplot(x='tip', data=tips, hue='day')
```

```python
sns.ecdfplot(x='tip', data=tips, stat='count', complementary=True)
```

```python
sns.ecdfplot(x='tip', data=tips, weights='tip')
plt.axhline(0.5, c='red')
```

```python
sns.ecdfplot(x='tip', data=tips, hue='day', palette='summer')
```

```python
p = sns.ecdfplot(x='tip', data=tips, hue='day', palette='summer', lw=3) #lw = line width
p.legend(['Thursday', 'Friday', 'Saturday', 'Sunday'], title='Day of Week', loc=4) #adjusting the graph manipulating on the created object 
```
