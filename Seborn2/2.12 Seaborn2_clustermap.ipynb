```python
import seaborn as sns 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
toy_data_dict = {
    'Name': ['Andy', 'Ben', 'Claire', 'David'],
    'Hours_Studied': [10, 20, 12, 28],
    'Score': [15, 20, 16, 25],
    'Street_Address': [11, 24, 18, 9]
}
```

```python
toy_data = pd.DataFrame(toy_data_dict)
toy_data.set_index('Name', inplace=True)

toy_data
```

```python
sns.clustermap(toy_data, figsize=(4, 4), annot=True) #how far away these columns are from each other, groups points that are close to each other 
```

```python
#HAC (Hierarchical agglomerative clustering)- is a method for grouping data points into clusters based on their similarity.
#Dendogram
```

```python
penguins = sns.load_dataset('penguins').dropna()
```

```python
penguins.head()
```

```python
penguins.shape
```

```python
numerical_cols = penguins.columns[2:6] #filtering only numerical columns

numerical_cols
```

```python
sns.clustermap(penguins[numerical_cols], standard_scale=1) #scaling data helps to notice more correlations 
```

```python
sns.clustermap(penguins[numerical_cols], standard_scale=1, method='single') #method allows to change the linkage
```

```python
sns.clustermap(penguins[numerical_cols], standard_scale=1, metric='cityblock') #metric allows to change the distance, default is Euclidan
```

```python
species_colors=penguins.species.map({
    'Adelie': 'blue',
    'Chinstrap': 'red',
    'Gentoo': 'green'
}) #assigning colors to different species
```

```python
len(species_colors) 
```

```python
species_colors[0:7]
```

```python
sns.clustermap(penguins[numerical_cols], standard_scale=1, row_colors=species_colors) #by doing so we can see how penguins were clustered
#can show importance of the categorical values
```
