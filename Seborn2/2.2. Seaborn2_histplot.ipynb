```python
import seaborn as sns
```

```python
sns.__version__ #checking current seaborn version
```

```python
penguins = sns.load_dataset("penguins")
```

```python
penguins.head()
```

```python
penguins.shape
```

```python
penguins.dropna(inplace=True)
penguins.shape
```

```python
sns.set_style('darkgrid')
```

```python
sns.histplot(penguins.bill_length_mm) #histogram for .bill_length_mm in dataframe penguins
```

```python
sns.histplot(y = "bill_length_mm", data=penguins) #horizontal histogram 
```

```python
sns.histplot(x = "bill_length_mm", data=penguins,
            kde=True) #kde plot 
```

```python
sns.histplot(x = "bill_length_mm", data=penguins, bins=20) #adjusting manually number of bins -> here 20, more granual view
```

```python
sns.histplot(x = "bill_length_mm", data=penguins,
             bins=[30,40,45,50,60]
            ) #building distinct bins
```

# Binwidth and binrange

```python
sns.histplot(x = "bill_length_mm", data=penguins,
             binwidth=5, binrange=(30, 60)
            ) #every bin is 5, starting from 30 and ending on 60 
```

# Statistics

```python
sns.histplot(x = "bill_length_mm", data=penguins,
             stat='count' #default stat used is count 
            )
```

```python
sns.histplot(x = "bill_length_mm", data=penguins,
             stat='density' #density- area of histogram sums to one 
            )
```

```python
sns.histplot(x = "bill_length_mm", data=penguins,
             stat='probability' #probability-height of histogram bars to one
            )
```

```python
sns.histplot(x = "bill_length_mm", data=penguins,
             stat='probability', fill=False, element='step',
             cumulative=True
            )
```

```python
sns.histplot(x = "bill_length_mm", data=penguins, hue='species') #showing the bill_length_mm in regards to species 
```

```python
sns.histplot(x = "bill_length_mm", data=penguins, 
             hue='species',
            element='step'
            ) #changing element to step allows to show the outlines of histograms in regards to the species 
```

```python
sns.histplot(x = "bill_length_mm", data=penguins, 
             hue='species',
            element='poly'
            ) #switching to polygons allows to see the overlap
```

```python
sns.histplot(x = "bill_length_mm", data=penguins, 
             hue='species',
            multiple='stack' #stacks number of peguins related to brackets on to of each other
            ) 
```

```python
sns.histplot(x = "bill_length_mm", data=penguins, 
             hue='species',
            multiple='fill' #shows proportion of each species for different bill_length
            )
```

# Bivariate Histplot

```python
sns.set_style('white')
```

```python
sns.histplot(x= "bill_length_mm", y='bill_depth_mm', data=penguins,
            cbar=True #shows legend 
            ) 
```

```python
sns.histplot(x= "bill_length_mm", y='bill_depth_mm', data=penguins,
             hue='species' #according to species, overlap not visible  
            )
```

```python
sns.histplot(x= "bill_length_mm", y='species', data=penguins,
             hue='species'
            )
```

# Styling

```python
sns.histplot(x='species', hue='sex', multiple='dodge', data=penguins
            )
```

```python
sns.histplot(x='species', hue='sex', multiple='dodge', data=penguins,
             shrink=0.8 #shrink allows to space the bars between different species 
            )
```

```python
sns.histplot(x='bill_length_mm', data=penguins,
            color='indigo'#changes color
            )
```

```python
sns.histplot(x='species', hue='sex', multiple='dodge', data=penguins,
             shrink=0.8,
             palette='bone'#changes color palette
            )
```

```python
sns.histplot(x='bill_length_mm', data=penguins,
            fill=False
            )
```
