```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg').dropna()

cars.shape
```

```python
cars.head(3)
```

```python
sns.set_style('whitegrid')
```

```python
cars.cylinders.value_counts()
```

```python
cars = cars[cars.cylinders.isin([4,6,8])] #filtering by number of cylinders
```

```python
sns.violinplot(cars.displacement)
```

```python
sns.violinplot(x=cars.cylinders, y=cars.displacement) #splits by category
```

```python
sns.violinplot(x='cylinders', y='displacement',
              data=cars
              ) #different syntax
```

```python
sns.violinplot(x='cylinders', y='displacement',
               hue='origin',
              data=cars
              ) #added category origin 
```

# Violinplot options 

```python
sns.violinplot(x='cylinders', y='displacement',
               hue='origin',
              data=cars[cars.origin.isin(['japan', 'europe'])]
              )
plt.legend(loc=2)
```

```python
sns.violinplot(x='cylinders', y='displacement',
               hue='origin',
              data=cars[cars.origin.isin(['japan', 'europe'])],
               split=True
              ) #split allows to show the spli data on the violinplot, works only when hue has 2 categories
plt.legend(loc=2)
```

```python
sns.violinplot(x='cylinders', y='displacement',
               hue='origin',
              data=cars[cars.origin.isin(['japan', 'europe'])],
               inner='quartiles'
              ) #changes the inner box into dash lines showing quartiles

plt.legend(loc=2)
```

```python
sns.violinplot(x='cylinders', y='displacement',
               hue='origin',
              data=cars[cars.origin.isin(['japan', 'europe'])],
               split=True,
               scale='count'
              ) #less information for european 6-cylinder than japanes ones, split first divides the data between 4 and 6 and then counts 

plt.legend(loc=2)
```

```python
sns.violinplot(x='cylinders', y='displacement',
               hue='origin',
              data=cars[cars.origin.isin(['japan', 'europe'])],
               split=True,
               scale='count',
               scale_hue=False
              ) #the data shows that 6 cylinders are less produced, data split dofferently than above 

plt.legend(loc=2)
```

```python
(
    cars[cars.origin.isin(['japan', 'europe'])]
    .groupby('cylinders')
    .origin
    .value_counts()
) #shows that the number of observations for 6 cylinders for both countries is low
```

# Styling 

```python
sns.violinplot(x='origin', y='displacement',
              order=['japan', 'europe', 'usa'],
               data=cars
              ) #ordering through violinplot
```

```python
sns.violinplot(x='origin', y='displacement',
              data=cars,
               linewidth=3
              ) #adjusting linewidth
```

```python
sns.violinplot(x='origin', y='displacement',
              data=cars,
               bw=0.2
              ) #adjusting bandwidth
```
