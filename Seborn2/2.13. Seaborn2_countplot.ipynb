```python
import seaborn as sns
```

```python
diamonds = sns.load_dataset('diamonds')

diamonds.head()
```

```python
diamonds = diamonds[diamonds.clarity.isin(['SI1', 'VS2'])]
```

```python
diamonds.shape
```

```python
sns.set_style('darkgrid')
```

```python
sns.countplot(x='color', data=diamonds)
```

```python
diamonds.color.value_counts(sort=False) #countplot is visual representation of value_counts
```

```python
sns.countplot(y='color', data=diamonds) #switches to horizontal by changing the axis
```

```python
diamonds.dtypes #pritnting the types of the variables
```

```python
diamonds.color.cat.categories #sorting alphabetically
```

```python
diamonds.cut.cat.categories
```

```python
color_order=['J', 'I', 'H', 'G', 'F', 'E', 'D']

sns.countplot(x='color', data=diamonds, order=color_order) #assigning manually the order by provided list
```

```python
diamonds.color.value_counts().index
```

```python
sns.countplot(x='color', data=diamonds,
              order=diamonds.color.value_counts().index)
```

```python
sns.countplot(x='color', data=diamonds,
              order=diamonds.color.value_counts().index[::-1]) #reversing the order 
```

```python
sns.countplot(x='color', data=diamonds, hue='clarity')
```

```python
sns.countplot(x='color', data=diamonds, hue='clarity', hue_order=['SI1', 'VS2']) #suplementing the hue_order with a list will allow to show only the categorical values we want
```

```python
sns.countplot(x='color', data=diamonds, color='skyblue') #assingning only one color
```

```python
sns.countplot(x='color', data=diamonds, palette='twilight')
```

```python
sns.countplot(x='color', data=diamonds,
             lw=4, ec='black', hatch='/') #hatch allows for pattern, lw for linewidth, ec for edgecolor
```
