```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
geyser = sns.load_dataset('geyser')
```

```python
geyser.head()
```

```python
geyser.shape
```

```python
sns.set_style('darkgrid')
```

```python
sns.jointplot(x=geyser.waiting, y=geyser.duration)
```

```python
sns.jointplot(x='waiting', y='duration', data=geyser)
```

```python
sns.jointplot(x='waiting', y='duration', data=geyser,
             kind='kde') #changing kind allows to change the type of plot displayed
```

```python
sns.jointplot(x='waiting', y='duration', data=geyser,
             kind='reg') #changing kind to regression
```

```python
sns.jointplot(x='waiting', y='duration', data=geyser,
             kind='hist') #changing kind to histogram
```

```python
g = sns.jointplot(x='waiting', y='duration', data=geyser)
```

```python
type(g)
```

```python
g.plot_joint(sns.kdeplot, color='grey', levels=5)
```

```python
sns.jointplot(x='waiting', y='duration', data=geyser,
             hue='kind') #adds different categories
```

```python
sns.jointplot(x='waiting', y='duration', data=geyser,
             hue='kind',
             kind='kde')
```

```python
sns.jointplot(x='waiting', y='duration', data=geyser,
             color='pink')
```

```python
sns.jointplot(x='waiting', y='duration', data=geyser,
             hue='kind', 
             palette='autumn')
```

```python
sns.jointplot(x='waiting', y='duration', data=geyser,
             height=8, 
             space=0.1,
             ratio=2) 
```

```python
sns.jointplot(x='waiting', y='duration', data=geyser,
             joint_kws={'marker': 5}) #changing marker
```

```python
sns.jointplot(x='waiting', y='duration', data=geyser,
             kind='reg',
             joint_kws={'ci': None, 'order': 2}) #results in curved line 
```

```python
sns.jointplot(x='waiting', y='duration', data=geyser,
             marginal_kws={'color': 'xkcd:golden'}) #changing only the color of the marginal plots
```

```python
sns.jointplot(x='waiting', y='duration', data=geyser,
              kind='kde',
             marginal_kws={'lw':4, 'fill': True}) #when switing to different plot remeber about changing the arguments of the magrinal kwywords matching the type of the plot
```
