```python
import pandas as pd
df = pd.read_csv("fb.csv",parse_dates=['Date'],index_col='Date')
df
```

```python
df.shift(1) #shifts the price by 1 day -> 8/15 now NaN
```

```python
df.shift(-1) #reverse action to shift(1)
```

```python
df['Prev Day Price'] = df['Price'].shift(1) #creates new column with name Prev Day Price while at the same time shifting the value of price by 1 day
df
```

```python
df['1 day change'] = df['Price'] - df['Prev Day Price'] #shows the difference in stock prices in relation to 1 day 
df
```

```python
df['5 day % return'] = (df['Price'] - df['Price'].shift(5))*100/df['Price'].shift(5) #shows the difference of % by 5 days
df
```

```python
df = df[['Price']] #returns back only 'Price' column
df
```

```python
df.index #frequency of this index is 'None'
```

```python
df.index = pd.date_range(start='2017-08-15',periods=10,freq='B') #freuquency now changed to business days
df.index
```
