```python
import pandas as pd
df1 = pd.DataFrame({
    "city": ["new york", "chicago", "orlando"],
    "temperature": [21,14,35],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["new york", "chicago", "orlando"],
    "humidity": [65,68,75],
})
df2
```

```python
df3 = pd.merge(df1,df2,on="city") #JOIN in SQL
df3
```

```python
df1 = pd.DataFrame({
    "city": ["new york", "chicago", "orlando", "baltimore"],
    "temperature": [21,14,35,32],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["new york", "chicago", "san francisco"],
    "humidity": [65,68,71],
})
df2
```

```python
df3 = pd.merge(df1,df2,on="city") #returns only values for cities that are in both dataframes (INNER JOIN), inner=default
df3
```

```python
df3 = pd.merge(df1,df2,on="city", how="outer") #returns values for all cities, putting NaN for missing information (OUTER JOIN)
df3
```

```python
df3 = pd.merge(df1,df2,on="city", how="left") #returns values for cities from first dataframe (LEFT JOIN)-> ignores san francisco
df3
```

```python
df3 = pd.merge(df1,df2,on="city", how="right") #returns values for cities from second dataframe (RIGHT JOIN)-> ignores baltimore, orlando
```

```python
df3 = pd.merge(df1,df2,on="city", how="outer",indicator=True) #indicator shows the dataframe from which the dta comes from
df3
```

```python
df1 = pd.DataFrame({
    "city": ["new york", "chicago", "orlando", "baltimore"],
    "temperature": [21,14,35,32],
    "humidity": [65,68,71,75]
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago", "new york", "san diego"],
    "temperature": [21,14,35],
    "humidity": [65,68,71]
})
df2
```

```python
df3 = pd.merge(df1,df2, on="city") #returns information for two temperatures and humidities that are for the same city in two dataframes
df3
```

```python
df3 = pd.merge(df1,df2, on="city", suffixes=('_left', '_right')) #suffixes add names for the header tables, compare with above 
df3
```
